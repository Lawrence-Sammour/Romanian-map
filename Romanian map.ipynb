```python
def Astar(startNode, heuristics, graph, goalNode="Bucharest"):
    priorityQueue = queue.PriorityQueue()
    distance = 0
    path = []

    priorityQueue.put((heuristics[startNode] + distance, [startNode, 0]))

    while priorityQueue.empty() == False:
        current = priorityQueue.get()[1]
        path.append(current[0])
        distance += int(current[1])

        if current[0] == goalNode:
            break

        priorityQueue = queue.PriorityQueue()

        for i in graph[current[0]]:
            if i[0] not in path:
                priorityQueue.put((heuristics[i[0]] + int(i[1]) + distance, i))

    return path
```

```python
def UCS(startNode, graph, goalNode="Bucharest"):
    priorityQueue = queue.PriorityQueue()
    distance = 0
    path = []

    priorityQueue.put((  distance, [startNode, 0]))

    while priorityQueue.empty() == False:
        current = priorityQueue.get()[1]
        path.append(current[0])
        distance += int(current[1])

        if current[0] == goalNode:
            break

        priorityQueue = queue.PriorityQueue()

        for i in graph[current[0]]:
            if i[0] not in path:
                priorityQueue.put((distance, i))

    return path
```

```python
def a_star_testing(start_point, Heuristics, distances, cooridnates):
  path_x_values = []
  path_y_values = []
  a = Astar(start_point, Heuristics, distances)
  keys = a
  values = list(map(cooridnates.get, keys))
  for i in range(len(values)):
    path_x_values.append(values[i][0]) 
    path_y_values.append(values[i][1])
  for i in range(len(a) - 1):
      plt.plot((path_x_values[i], path_x_values[i+1]), ((path_y_values[i]), (path_y_values[i+1])), color = "blue")
```

```python
def UCS_testing(start_point,  distances, cooridnates):
  path_x_values = []
  path_y_values = []
  a = UCS(start_point, distances)
  keys = a
  values = list(map(cooridnates.get, keys))
  for i in range(len(values)):
    path_x_values.append(values[i][0]) 
    path_y_values.append(values[i][1])
  for i in range(len(a) - 1):
      plt.plot((path_x_values[i], path_x_values[i+1]), ((path_y_values[i]), (path_y_values[i+1])), color = "green")
```
